# Q1
Use web scarping with Python to extract all faculty and staff information from CSUF catalog
at https://catalog.fullerton.edu/content.php?catoid=52&navoid=6265

```python
import requests
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
url = 'https://catalog.fullerton.edu/content.php?catoid=52&navoid=6265'
webpage = requests.get(url)
soup = BeautifulSoup(webpage.text,'html.parser')
blocks = soup.find(class_='block_content')

 

rows = blocks.find_all('p')

 

Faculty_staff_containers = []  

for row in rows:

    strong_id = row.find_all('strong')
  
    link_id = row.find_all('a')

    if strong_id  and not link_id:

        Faculty_staff_containers.append(row.get_text())
        
        
print(Faculty_staff_containers)
len(Faculty_staff_containers)
```

# Q4
Compute the number of new faculty members and the number of new staff members for each
year in the period of 2010-2019. Use a bar plot to present your results. In the bar plot, the
horizontal axis should be year 2010, 2011, …., 2019. Use different colors to distinguish the bars
for faculty and staff. See the link

To filter position of faculty having lecturer/professor from 2010-2019

```python
#LecPro gives all the faculty(lecturer/professor) from the start till 2019
LecPro = FacultyStaff[FacultyStaff['Position'].str.contains('Lecturer|Professor')]
#LecPro
#sort_year_wise gives data having year from 2010-2019
sort_year_wise = LecPro[LecPro['Year'].str.contains('2010|2011|2012|2013|2014|2015|2016|2017|2018|2019')]
#sort_year_wise
Group_by_year = sort_year_wise.groupby('Year', as_index=False).agg({"Position":"count"})
Group_by_year
```

To filter position of staff having Lecturer/professor from 2010-2019

```python
#LecPro2 gives all the staff(lecturer/professor) from the start till 2019
LecPro2 = FacultyStaff[~FacultyStaff['Position'].str.contains('Lecturer|Professor')]
#LecPro2
#sort_year_wise2 gives data having year from 2010-2019
sort_year_wise2 = LecPro2[LecPro2['Year'].str.contains('2010|2011|2012|2013|2014|2015|2016|2017|2018|2019')]
#sort_year_wise2
Group_by_year2 = sort_year_wise2.groupby('Year', as_index=False).agg({"Position":"count"})
Group_by_year2
```

Merging both the dataframes(faculty & staff) on year

```python
#merging the dataframes (faculty count and staff count)
result = pd.merge(Group_by_year, Group_by_year2, on = 'Year')
#result
#renaming the column names to faculty_count and staff_count
result.rename(columns={'Position_x':'Faculty_count','Position_y':'Staff_count'}, inplace=True)
result
```

Ploting bar chat for faculty count and staff count over year

```python
result.plot(x='Year', y=['Faculty_count','Staff_count'], kind='bar')
```

# Finding
The faculty count(lecturer or professor) and the staff count for the year 2015 is highest having 72 and 51. If seen in 2010 staff_count is 3 very less in comparison to faculty_count 23.

# Q5 
For each year in the period of 2010-2019, compute the number of new faculty or staff
members having Ph.D. or J.D. degree. Then use a bar plot to present your results. In the bar plot,
the horizontal axis should be year 2010, 2011, …., 2019. Explain any findings from the bar plot.

To filter degree (Ph.D and J.D.) from FacultyStaff from 2010-2019

```python
#Degree_count gives all the faculty or staff having Ph.D/J.D. from 2010 till 2019
Degree_count = FacultyStaff[FacultyStaff['Degrees'].str.contains('Ph.D.|J.D.')]
sort_degree_count = Degree_count[Degree_count['Year'].str.contains('2010|2011|2012|2013|2014|2015|2016|2017|2018|2019')]
sort_degree_count
```

Getting the count of the degrees and renaming the column name to Degrees_count

```python
#Degree_count_total gives the count of degrees and renaming the column name
Degree_count_total = sort_degree_count.groupby('Year', as_index=False).agg({"Degrees":"count"})
Degree_count_total.rename(columns={'Degrees':'Degrees_count'}, inplace=True)
Degree_count_total
```

Ploting Bar chart for the degree count over the years

```python
Degree_count_total.plot(x='Year', y=['Degrees_count'], kind='bar')
```

# Finding
The trend from the bar chart shows that 2015 has the highest degrees (62 degress) for a staff/faculty and 2010 has the lowest number (19). The number of degrees kept reducing over the years from 2015 to 2019.
